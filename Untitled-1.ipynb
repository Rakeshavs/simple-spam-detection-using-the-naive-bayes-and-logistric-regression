```python
# importing all the libraries
import pandas as pd # for data manipulation
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
# nltk library 
nltk.download('stopwords')

from nltk.corpus import stopwords

```

```python
#importing the dataset 

import pandas as pd

# Load Enron dataset
df = pd.read_csv(
    r"C:\Users\medha\Downloads\enron_spam_data\enron_spam_data.csv",
    encoding="latin-1"
)

# Combine Subject + Message as text
df["text"] = df["Subject"].fillna("") + " " + df["Message"].fillna("")

# Select only required columns
df = df[["Spam/Ham", "text"]]

# Renaming columns
df.columns = ["label", "text"]

print(df.head())
print(df["label"].value_counts())

```

```python
stop_words = set(stopwords.words("english"))
# lowering the text ,and allowing the alphabet , removing the stop words

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)
    words = text.split()
    words = [word for word in words if word not in stop_words]
    return " ".join(words)

df["clean_text"] = df["text"].apply(clean_text)
```

```python
#creating the new columsn of o and 1 for spam
df["label"] = df["label"].map({"ham": 0, "spam": 1})
```

```python
#train test split the cleaned dataset 
X_train, X_test, y_train, y_test = train_test_split(
    df["clean_text"], df["label"], test_size=0.2, random_state=42
)
```

```python
# TF-IDF Vectorization
tfidf = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1,2))

X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)
```

```python
#naive bayes
nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)
```

```python
#logistic regression
lr = LogisticRegression(C=2.0, max_iter=1000)

lr.fit(X_train_tfidf, y_train)
```

```python
# evalution metrics
models = {
    "Naive Bayes": nb,
    "Logistic Regression": lr
}

results = []

for name, model in models.items():
    y_pred = model.predict(X_test_tfidf)
    
    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred)
    })

results_df = pd.DataFrame(results)
print(results_df)
```

```python
#graphical representation of bar graph
results_df.set_index("Model").plot(kind="bar", figsize=(10,5))
plt.title("Model Comparison")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.show()
```

```python

# confusion matrix for logistic regression
cm = confusion_matrix(y_test, lr.predict(X_test_tfidf))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

```python
# confusion matrix for naive bayes
cm = confusion_matrix(y_test, nb.predict(X_test_tfidf))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix - naive bayes")
plt.show()
```

```python
#predicitng a sample
def predict_spam(text):
    cleaned = clean_text(text)
    vector = tfidf.transform([cleaned])
    prediction = lr.predict(vector)[0]
    return "Spam" if prediction == 1 else "Not Spam"

print(predict_spam("“Lowest prices on medicines! Order now and get 70% off. Call 1800-XXX-XXX.”"))
print(predict_spam("Are we meeting today at 5 PM?"))
```
